```python
#create a data frame
benchmarks = funcs.getAllBenchmarks('../benchmarks')
cols = ['total','ground','solve']
rows = ['BM{}'.format(i+1) for i, bm in enumerate(benchmarks)]
dct = {}
for c in cols:
    dct[c] = [0.0]*len(rows)
df = pd.DataFrame(dct, index=rows)
#df.to_csv('../benchmark_data/' + encoding + '.csv')
```

```python
# reading json benchmark data
files = glob.glob("../benchmark_data/*det*/*.json")
bmdfs = []
for file in files:
    bmdfs.append(BMDataFormat())
    bmdfs[-1].load(file)
```

```python
# insert benchmark data into data frame
for bmdf in bmdfs:
    df.loc[df.index[int(bmdf.data['instance'][0:2])-1]] = [
                  bmdf.data['statistics']['total'],
                  bmdf.data['statistics']['groundingTime'],
                  bmdf.data['statistics']['solvingTime']
                 ]
df.to_csv('../benchmark_data/merger_ps_rec4_small+merger_w_det2/merger_ps_rec4_small+merger_w_det2.csv')
df
```

```python
from merger import Merger
# initialize merger
merger = Merger("../encodings/sequential", "../benchmark_data")
def merge(benchmark_id):
        benchmark = benchmarks[benchmark_id-1]
        return merger.merge(benchmark, vizualize=True, save_data=True, small_switcher=True, deterministic_waiter=False, check_validity=False)
        
def mergeAll(exclude=[22,23,26]):
    for i in range(len(benchmarks)):
        benchmark_id = i+1
        benchmark = benchmarks[i]
        if benchmark_id in exclude:
            continue
        merger.merge(benchmark, vizualize=False, save_data=True, automated=False, small_switcher=True, deterministic_waiter=True, check_validity=False)
```
